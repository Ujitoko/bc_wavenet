```python
# 指定されたタイムステップで時間足を取得。
def get_json(num_time_steps):
    rand_toTs = np.random.randint(low=1506970800 + num_time_steps, high=1521370800)
    #print(rand_toTs)
    url = 'https://min-api.cryptocompare.com/data/histohour?fsym=BTC&tsym=JPY&limit={0}&e=bitFlyer&toTs={1}'.format(num_time_steps, rand_toTs)
    print(url)
    response = urllib.request.urlopen(url)
    content = json.loads(response.read().decode('utf8'))
    return content
```

```python
# batch内の1 dataを作成
def make_single_batch(num_time_steps):
    hours_content = get_json(num_time_steps)
    single_batch = []
    for data in hours_content["Data"]:
        tmp_data = [data["close"], data["high"], data["low"], data["open"], data["volumefrom"], data["volumeto"]]
        single_batch.append(tmp_data)
        #print(data["time"])
    return single_batch
```

```python
# 正規化
def min_max_normalize(x, axis=None):
    min = x.min(axis=axis, keepdims=True)
    max = x.max(axis=axis, keepdims=True)
    result = (x-min)/(max-min)
    return result
```

```python
# batchを作成
def make_multi_batch(num_batch_size, num_time_steps):
    multi_batch = []
    for i_batch in range(num_batch_size):
        multi_batch.append(make_single_batch(num_time_steps))
        
    # 正規化
    multi_batch_np = np.array(multi_batch)
    norm_multi_batch_np = min_max_normalize(multi_batch_np, axis=1)
    print()
    print(multi_batch_np)
    print()
    print(norm_multi_batch_np)

    return np.array(norm_multi_batch_np)
```
